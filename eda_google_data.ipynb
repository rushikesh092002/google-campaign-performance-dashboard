```python

df['Cost'] = df['Cost'].replace('[\$,]', '', regex=True).astype(float)

```

```python
df['Sale_Amount'] = df['Sale_Amount'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df.head()
```

```python
df.info()
```

```python

df['Conversion Rate (Recalc)'] = df['Conversions'] / df['Clicks'].replace(0, np.nan)

df['Conversion Rate'] = df['Conversion Rate'].fillna(df['Conversion Rate (Recalc)'])
df.drop(columns=['Conversion Rate (Recalc)'], inplace=True)

```

```python
df.info()
```

```python
missing_data = df.isnull().sum().sort_values(ascending=False)
print(missing_data)
```

```python
df['Conversion Rate'] = df['Conversion Rate'].fillna(df['Conversions'] / df['Clicks'].replace(0, np.nan))
```

```python
missing_data = df.isnull().sum().sort_values(ascending=False)
print(missing_data)
```

```python
df = df.dropna(subset=['Conversions'])
```

```python
avg_rev_per_conv = df['Sale_Amount'].sum() / df['Conversions'].sum()
avg_cpc = df['Cost'].sum() / df['Clicks'].sum()
avg_ctr = df['Clicks'].sum() / df['Impressions'].sum()
```

```python
df['Sale_Amount'] = df['Sale_Amount'].fillna(df['Conversions'] * avg_rev_per_conv)
df['Cost'] = df['Cost'].fillna(df['Clicks'] * avg_cpc)
df['Impressions'] = df['Impressions'].fillna(df['Clicks'] / avg_ctr)
```

```python
print(df.isnull().sum())
```

```python
df['Clicks'] = df['Clicks'].fillna(df['Cost'] / avg_cpc)
```

```python
df['Impressions'] = df['Impressions'].fillna(df['Impressions'].median())
```

```python
df['Conversion Rate'] = df['Conversion Rate'].fillna(df['Conversion Rate'].mean())
```

```python
print(df.isnull().sum())
```

```python
df['Cost']=df['Cost'].fillna(df['Cost'].mean())
```

```python
df.head()
```

```python


# Calculate floor mean and convert to float with .0
mean_val = np.floor(df['Leads'].mean())
clean_val = float(mean_val)


```

```python
clean_val
```

```python
# Fill missing values with that
df['Leads'] = df['Leads'].fillna(clean_val)
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
df['CTR'] = df['Clicks'] / df['Impressions']
df['CPC'] = df['Cost'] / df['Clicks'].replace(0, np.nan)
df['CPA'] = df['Cost'] / df['Conversions'].replace(0, np.nan)
```

### ✅ 1. Total Clicks by Campaign ###

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.barplot(x='Campaign_Name', y='Clicks', data=df, estimator='sum', ci=None)
plt.xticks(rotation=45)
plt.title('Total Clicks by Campaign')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Impressions', y='Clicks', hue='Campaign_Name', data=df)
plt.title('Impressions vs Clicks')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Cost'], bins=20, kde=True)
plt.title('Cost Distribution')
plt.xlabel('Cost')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Device', y='Leads', data=df)
plt.title('Lead Distribution by Device')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='Campaign_Name', y='Conversion Rate', data=df, estimator='mean', ci=None)
plt.xticks(rotation=45)
plt.title('Average Conversion Rate by Campaign')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=df[['Leads', 'Conversions']])
plt.title('Leads vs Conversions Trend')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Campaign_Name', y='Cost', data=df, color='red', label='Cost')
sns.barplot(x='Campaign_Name', y='Sale_Amount', data=df, color='green', label='Sales')
plt.xticks(rotation=45)
plt.title('Cost vs Sales by Campaign')
plt.legend()
plt.tight_layout()
plt.show()
```
